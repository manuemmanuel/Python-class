**`Logical Operators`**

Logical operators are used to check whether an expression is True or False. They are used in decision-making.

image.png

````python
# Initial values
number1 = 10
number2 = 5
number3 = 15

# Using 'and' operator
print("Is number1 greater than number2 and number1 less than number3? ", (number1 > number2) and (number1 < number3))
# True (because both conditions are True)

# Using 'or' operator
print("Is number1 greater than number2 or number1 greater than number3? ", (number1 > number2) or (number1 > number3))
# True (because one condition is True)

# Using 'not' operator
print("Is it not true that number1 is greater than number3? ", not (number1 > number3))
# True (because number1 is not greater than number3)
````

**Bitwise operators**


Bitwise operators act on operands as if they were strings of binary digits. They operate bit by bit, hence the name.

In the table below: Let x = 10 (0000 1010 in binary) and y = 4 (0000 0100 in binary)


image.png




````python
# Initial values
number1 = 10  # In binary: 1010
number2 = 4   # In binary: 0100
# Bitwise AND
result = number1 & number2
print(f"Bitwise AND of {number1} & {number2} = {result}")  # 1010 & 0100 = 0000 (result = 0)
# Bitwise OR
result = number1 | number2
print(f"Bitwise OR of {number1} | {number2} = {result}")   # 1010 | 0100 = 1110 (result = 14)
# Bitwise XOR
result = number1 ^ number2
print(f"Bitwise XOR of {number1} ^ {number2} = {result}")  # 1010 ^ 0100 = 1110 (result = 14)
# Bitwise NOT
result = ~number1
print(f"Bitwise NOT of ~{number1} = {result}")             # ~1010 = -1011 (result = -11)
# Bitwise Left Shift
result = number1 << 2
print(f"Left Shift of {number1} << 2 = {result}")          # 1010 << 2 = 101000 (result = 40)
# Bitwise Right Shift
result = number1 >> 2
print(f"Right Shift of {number1} >> 2 = {result}")         # 1010 >> 2 = 0010 (result = 2)
````

**Identity operators**

In Python, **is** and **is not** are used to check if two values are located at the same memory location.

It's important to note that having two variables with equal values doesn't necessarily mean they are identical.

image.png


````python
# Example of identity operators

a = [1, 2, 3]
b = [1, 2, 3]
c = a
print(type(a))
print(type(b))
print(type(c))
# Using 'is' operator
print(a is b)  # False, because 'a' and 'b' are different objects in memory, even though they have the same value
print(a is c)  # True, because 'a' and 'c' point to the same object in memory

# Using 'is not' operator
print(a is not b)  # True, because 'a' and 'b' are different objects
print(a is not c)  # False, because 'a' and 'c' point to the same object
````

````python
# Example of identity operators with strings

str1 = "helloooooooooooooooooooooooooooooooooo"
str2 = "helloooooooooooooooooooooooooooooooooo"
str3 = str1
# Using 'is' operator
print(str1 is str2)  # True, because Python reuses the same object for small strings (interning)
print(str1 is str3)  # True, because 'str3' refers to the same object as 'str1'
# Using 'is not' operator
print(str1 is not str2)  # False, because 'str1' and 'str2' point to the same object (interning)
print(str1 is not str3)  # False, because 'str1' and 'str3' are the same object
````

**String interning** is an optimization technique used in Python, particularly for small and frequently used strings, to avoid unnecessary memory usage and speed up comparisons. When Python interns a string, it stores a single copy of that string in memory, and all references to that string point to the same memory location.

**3. Branching Statements**

    a. break
       Terminates a loop early when a specified condition is satisfied..



````python
for i in range(10):
    if i == 8:
        break
    print(i)  # Outputs 0, 1, 2, 3, 4,5,6,7,
````

 b. continue

      Skips the rest of the loop body and starts the next iteration.

````python
for i in range(5):
    if i == 2:
        continue  #start the next iteration
    print(i)  # Outputs 0, 1, 3, 4
````

    How It Works
    The loop starts, iterating through the numbers in range(5) → [0, 1, 2, 3, 4].
    For each value of i:
    If i equals 2, the continue statement is executed.
    This skips the print(i) statement for this iteration and moves to the next number in the loop.
    For other values of i, the print(i) statement is executed.

    c. pass
    
    Acts as a placeholder where code will be added later. It does nothing.

````python
x = 2
if x > 5:
    pass  # No action, placeholder
else:
    print("x is not greater than 5")
````

Syntax of match-case


     match variable:
        case value1:
        # Code block for value1
        case value2:
        # Code block for value2
        case _:
        # Default case (like "else")


Example 1: Simple Switch-Case with match-case
Scenario: Determine the day of the week

````python
while True:
    day_number = int(input("Enter a day number (1-7) or 0 to exit: "))

    if day_number == 0:  # Exit condition
        print("Exiting the program. Goodbye!")
        break

    match day_number:
        case 1:
            print("Monday")
        case 2:
            print("Tuesday")
        case 3:
            print("Wednesday")
        case 4:
            print("Thursday")
        case 5:
            print("Friday")
        case 6:
            print("Saturday")
        case 7:
            print("Sunday")
        case _:
            print("Invalid day number! Please enter a number between 1 and 7.")
````

**Case Study: Online Movie Ticket Booking System**


This case study will detail the design and flow of an Online Movie Ticket Booking System, showcasing its capabilities for handling movie selection, showtime choices, ticket quantities, ticket types (normal or premium), and applying various discounts.

**System Features and Flow:**
1. Movie Selection:

The user is presented with a list of available movies, each associated with a unique number (e.g., 1 for Action, 2 for Drama, and 3 for Comedy).
The user selects a movie by entering the number corresponding to their choice.

2. Ticket Selection:

After selecting the movie, the user is prompted to enter the number of tickets they wish to book.
The system offers two types of tickets: Normal and Premium, each with different prices.
The user selects the ticket type, which determines the cost per ticket.

3. Showtime Selection:

A set of predefined showtimes (in 24-hour format) is offered, including options like 10:00 AM, 2:00 PM, 6:00 PM, and 10:00 PM.
The user selects a showtime by entering the corresponding number.

4. Discount Application:

Applies discounts based on conditions:
(10% discount if more than 5 tickets are booked.)

5. Total Cost Calculation:

After applying any applicable discounts, the system calculates the total cost of the booking.
The user is then presented with a summary of their booking: movie, number of tickets, ticket type, showtime, day of the week, and the total cost.

6. Loop for Multiple Bookings:

The system asks the user if they wish to make another booking.
If the user chooses "yes", the system repeats the entire process. If "no", the system ends the booking session and thanks the user.

````python
from datetime import datetime
# Main program loop
while True:
    # Display available movies and get user's choice
    print("Available Movies:")
    print("1. Action")
    print("2. Drama")
    print("3. Comedy")

    # Get the movie choice from user
    movie_choice = int(input("Choose a movie (1-3): "))

    # Get the number of tickets and ticket type
    number_of_tickets = int(input("Enter the number of tickets: "))
    print("Ticket Types: 1. Normal ($10), 2. Premium ($20)")
    ticket_type = int(input("Choose ticket type (1 for Normal, 2 for Premium): "))

    # Display available showtimes as options
    print("Available Showtimes (in 24-hour format):")
    print("1. 10:00 AM")
    print("2. 14:00 PM (2 PM)")
    print("3. 18:00 PM (6 PM)")
    print("4. 22:00 PM (10 PM)")

    show_time_choice = int(input("Choose a showtime (1-4): "))

    # Map the user's choice to the corresponding showtime
    if show_time_choice == 1:
        show_time = 10
    elif show_time_choice == 2:
        show_time = 14
    elif show_time_choice == 3:
        show_time = 18
    elif show_time_choice == 4:
        show_time = 22
    else:
        print("Invalid choice! Defaulting to 10:00 AM.")
        show_time = 10  # Default to 10 AM if invalid input

    # Get current day of the week
    current_day = datetime.now().strftime("%A")  # Get current day name

    # Determine ticket price
    if ticket_type == 1:  # Normal Ticket
        ticket_price = 10
    elif ticket_type == 2:  # Premium Ticket
        ticket_price = 20

    # Calculate total cost
    total_cost = ticket_price * number_of_tickets

    # Initialize discount variable
    discount = 0

    # Apply discount for more than 5 tickets
    if number_of_tickets > 5:
        discount = 0.10  # 10% discount for more than 5 tickets

    # Apply the discount
    total_cost -= total_cost * discount  # Apply the discount to total cost (total_cost = total_cost - (total_cost * discount))

    # Display the booking summary
    if movie_choice == 1:
        movie_name = "Action"
    elif movie_choice == 2:
        movie_name = "Drama"

    elif movie_choice == 3:
        movie_name = "Comedy"

    print("\n--- Booking Summary ---")
    print(f"Movie: {movie_name}")
    print(f"Number of Tickets: {number_of_tickets}")
    print(f"Ticket Type: {'Normal' if ticket_type == 1 else 'Premium'}")
    print(f"Showtime: {show_time}:00")
    print(f"Day: {current_day}")
    print(f"Total Cost: ${total_cost:.2f}")

    # Ask if the user wants to continue booking
    continue_booking = input("\nDo you want to book another ticket? (yes/no): ").strip().lower()
    if continue_booking != "yes":
        print("Thank you for using the Online Movie Ticket Booking System. Goodbye!")
        break
````

#**Lazy evaluation**

Lazy evaluation, also known as call-by-need, is a programming strategy where expressions are not evaluated until their values are actually required. It is commonly used to improve performance and manage resource consumption efficiently.

#1. Using Logical Operators

Logical operators (and, or) inherently exhibit lazy evaluation.


````python
x = 10
y = 0

# Lazy evaluation with `and`
if y != 0 and (x / y > 1):  # The second condition won't be evaluated as `y != 0` is False
    print("Condition met")
else:
    print("Division skipped due to lazy evaluation")

# Lazy evaluation with `or`
if x > 5 or y == 0:  # The second condition (y == 0) won't be evaluated because x > 5 is True
    print("First condition is True, so no need to check the second condition")
````

#2. Generators

Generators in Python are a special type of iterable, used to generate values on the fly without storing them in memory. They are useful for working with large datasets or creating infinite sequences where generating all the values at once would be inefficient or impossible.

Generators in Python are inherently lazy. They produce items one at a time and compute the next value only when explicitly requested.

````python
gen = (x * 2 for x in range(5))  # Generator expression

# Values are computed only when accessed
print(next(gen))  # Output: 0
print(next(gen))  # Output: 2
print(next(gen))  #
print(next(gen))
print(next(gen))

````

In Python, next() is a built-in function used to retrieve the next value from an iterator, such as a generator.

# zip function

The zip function in Python combines multiple iterables element by element, creating an iterator of tuples. It works lazily, meaning it generates the tuples on demand rather than creating the entire list in memory at once. This makes it efficient for handling large iterables.

      Syntax:
          zip(iterable1, iterable2, ...)

````python
list1 = [1, 2, 3]
list2 = ['a', 'b', 'c']
zipped = zip(list1, list2)

print(list(zipped))  # Output: [(1, 'a'), (2, 'b'), (3, 'c')]
````

The zip function is considered lazy because it does not generate all the tuples in memory upfront. Instead, it produces each tuple on demand, one at a time, as you iterate over the zip object. This behavior is particularly useful when working with large or infinite iterables, as it avoids consuming unnecessary memory.

#itertools

The itertools module provides tools for lazy iteration, generating items on demand to save memory and handle large or infinite datasets efficiently.



#Infinite Iterators:

1. count(start=0, step=1): Generates numbers endlessly.

````python
from itertools import count
for i in count(10, 2):
    if i > 20:
        break
    print(i)  # Output: 10, 12, 14, 16, 18, 20
````

2. cycle(iterable): Repeats an iterable indefinitely.

````python
from itertools import cycle
colors = cycle(['red', 'green'])
for _ in range(4):
    print(next(colors))  # Output: red, green, red, green
````

3. repeat(value, times=None): Repeats a value a fixed number of times.




````python
from itertools import repeat
print(list(repeat('hello', 3)))  # Output: ['hello', 'hello', 'hello']
````

# **How Python Works: A Detailed Overview**

Python is an interpreted, high-level, and dynamically typed language. Here’s a breakdown of how Python works under the hood:

#1. Python Code Execution Flow


**Step 1: Writing Python Code**

You write Python code in a text editor or Integrated Development Environment (IDE) (e.g., example.py).


**Step 2: Python Interpreter**

When you run the Python program, the Python interpreter takes the code and processes it in a series of steps. The interpreter itself is a program that reads and executes Python code.


Interactive Mode:

You can interact with Python directly via the terminal or command prompt by typing Python commands interactively.

Script Mode:

You execute Python files by running them as scripts using the python command
(e.g., python example.py).

#2. Steps in Python Code Execution

**Lexical Analysis:**

The Python interpreter first reads the code and performs tokenization, breaking the source code into smaller units called tokens (e.g., keywords, variables, operators).


Example: In the line x = 5 + 3, it splits it into tokens: x, =, 5, +, 3.


**Parsing:**

Python uses a parser to convert the tokens into an Abstract Syntax Tree (AST), a tree-like structure that represents the syntactic structure of the program.
The AST makes it easy for the interpreter to understand the program's flow and logic.


**Bytecode Compilation:**

The AST is then compiled into bytecode. This bytecode is a lower-level, platform-independent representation of your code.
Python compiles the source code into a .pyc (Python Compiled) file, which is
stored in the    _ pycache _ directory.
This allows for faster execution when the program is run again, as the Python interpreter can reuse the bytecode instead of recompiling it.


**Python Virtual Machine (PVM):**

The Python Virtual Machine (PVM) is responsible for interpreting the bytecode and executing it on the actual machine.
The PVM reads the bytecode instructions and translates them into actions on the computer’s hardware.


3. Dynamic Typing

Python is a dynamically typed language, meaning:


You don’t need to declare the type of a variable explicitly. The interpreter figures it out at runtime based on the value assigned.

    x = 5     # x is an integer
    x = "Hello"  # x is now a string

This makes Python flexible but also means that types are checked during execution, not at compile time.
